```python
import torch

inputs = torch.tensor(
  [[0.43, 0.15, 0.89], # Your     (x^1)
   [0.55, 0.87, 0.66], # journey  (x^2)
   [0.57, 0.85, 0.64], # starts   (x^3)
   [0.22, 0.58, 0.33], # with     (x^4)
   [0.77, 0.25, 0.10], # one      (x^5)
   [0.05, 0.80, 0.55]] # step     (x^6)
)
```

```python
inputs.shape
```

```python
input_query = inputs[1]
input_query
```

```python
attn_scores_2 = torch.empty(inputs.shape[0])
for idx , val in enumerate(inputs):
    attn_scores_2[idx] = torch.dot(val,input_query)
print(attn_scores_2)
    
```

```python
attn_scores_temp = attn_scores_2/attn_scores_2.sum()
print("Normalized weights:", attn_scores_temp)
print(attn_scores_temp.sum())
```

```python
attn_scores_final = torch.softmax(attn_scores_2,dim =0)
```

```python
input_query.shape
```

```python
context_vec = torch.zeros(input_query.shape)
for i, val in enumerate(inputs):
    context_vec += attn_scores_final[i]*inputs[i]

print(context_vec)
```
